### Automatic differentiation using PyTorch’s Autograd system.

We'll be trying to understand what exactly is going on here... :D 

Autogradient, which is the automatic differentiation methodology used in Torch, is a neat and efficient way to calculate the gradient of the outcome wrt to the various various depends on. We'll be going through a quick example

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(4.0, requires_grad=True)

c = (a**2) + (b**2) +(2*a*b)
```

```python
print(c.backward())
print(a.grad) ## 2a + 2b 
print(b.grad) ## 2b + 2a
```

Now, that is the basic way of seeing how things works. In Neural Networks you use a Optimizier that uses these gradients to update the weights.

Consider now, that instead of the above, c is a final outcome from a combination of given like: $ W_a*a + W_b*b + bias$. This is the traditional Neural Network concept where to inputs are fed into one Artificial Neuron.

Going Deep into it, we can describe what is going under the hood (for autograd). Given the final outcome c in this example, assume c' is the actual values that needed to be attained. So there would be a Loss variable defined wrt to come constant c' and some values c. 

$$ L = f(c) $$

This is how the backward works now, the gradient of c is generated which is: $ \frac{\partial L}{\partial c} $. Now remember that we also need to update $ W_a*a $ & $ W_b*b$. And this is pretty simple according to the chain rule:
$$ W_a = \frac{\partial L}{\partial a} = \frac{\partial L}{\partial c} * \frac{\partial c}{\partial a} $$
$$ W_b = \frac{\partial L}{\partial b} = \frac{\partial L}{\partial c} * \frac{\partial c}{\partial b} $$

In particular, $ \frac{\partial c}{\partial a} $ and $ \frac{\partial c}{\partial b} $ are stored as rules, which in our examples would be $W_a^*$ and $W_b^*$. This is the Jacobian Matrix which forms [$W_a^*$, $W_b^*$]. Now the equations are resolved to:
$$ W = Jacobian * \frac{\partial L}{\partial c}$$

This my friends, is the reflection of the autograd of a simple system.

## 3) Write a program to implement AND OR gates using Perceptron.

We are working with 3 Weights: Coefficients Wx, Wy and Bias Wo. We Work on the Perceptron logic, consider weights to be integer values and not float values and let alpha learning to 1 instead of float values

```python
print("x u y")
Input = [[0,0], [0,1], [1,0], [1,1]]
Wx = 0
Wy = 0
Wo = 0
Output = [False, False, False, False]

while all(Output) != True:
    for n in range(len(Input)):
        x = Input[n][0]
        y = Input[n][1]

        Solution = x | y
        Ans = Wo + x*Wx + y*Wy #Sigma Function


        Prediction = 1 if Ans >= 1 else 0 #Activation Function

        if Prediction != Solution:
            Output[n] = False

            delta = Solution - Prediction

            Wo += delta
            Wx += delta * x
            Wy += delta * y
        else:
            Output[n] = True

print(Output)

if all(Output) == True:
    print(Wo, Wx, Wy)
    for item in Input:
        print("x =", item[0], "y =", item[1], "Output =", item[0] | item[1], "Solution =",1 if (Wo + item[0]*Wx + item[1]*Wy) else 0)
```

```python
print("x n y")
Input = [[0,0], [0,1], [1,0], [1,1]]
Wx = 0
Wy = 0
Wo = 0
Output = [False, False, False, False]

while all(Output) != True:
    for n in range(len(Input)):
        x = Input[n][0]
        y = Input[n][1]

        Solution = x & y
        Ans = Wo + x*Wx + y*Wy #Sigma Function


        Prediction = 0 if Ans <= 0 else 1 #Activation Function

        if Prediction != Solution:
            Output[n] = False

            delta = Solution - Prediction

            Wo += delta
            Wx += delta * x
            Wy += delta * y
        else:
            Output[n] = True

print(Output)

if all(Output) == True:
    print(Wo, Wx, Wy)
    for item in Input:
        print("x =", item[0], "y =", item[1], "Output =", item[0] & item[1], "Solution =", 0 if (Wo + item[0]*Wx + item[1]*Wy) <= 0 else 1)
```

## 4) Implementation of XOR Problem using PyTorch Neural Network.

```python
import torch.nn as nn
```

```python
Label_XOR = torch.tensor([[0.],[1.],[1.],[0.]])
Q = torch.tensor([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
```

We have to recall that XOR actually requires two lines of separation and cannot be just identified with a input -> neuron, but requires another additional scrutiny.

```python
class XOR(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(2, 2)   # hidden layer
        self.output = nn.Linear(2, 1)   # output layer

    def forward(self, x):
        x = self.hidden(x)
        x = torch.sigmoid(x)
        x = self.output(x)
        return x
```

```python
model = XOR()
criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.5)
```

```python
epochs = 5000

for epoch in range(epochs):
    # 1. Forward pass
    y_pred = model(Q)

    # 2. Compute loss
    loss = criterion(y_pred, Label_XOR)

    # 3. Backward pass
    optimizer.zero_grad()
    loss.backward()

    # 4. Update weights
    optimizer.step()

    # Logging
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    logits = model(Q)
    preds = (logits > 0).float()

    print("Logits:\n", logits)
    print("Predictions:\n", preds)
    print("Targets:\n", Label_XOR)
```

## 5) Implement Simple below Neural Network to solve regression problem.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('house_price_full+(2) - house_price_full+(2).csv')

x = df[['bedrooms', 'sqft_living']].values.astype('float32')
y = df[['price']].values.astype('float32')

print(type(x))
```

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()
```

```python
x_scaled = scaler_x.fit_transform(x)
y_scaled = scaler_y.fit_transform(y)
```

```python
x_t = torch.tensor(x_scaled, dtype=torch.float32)
y_t = torch.tensor(y_scaled, dtype=torch.float32)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 2),
            nn.ReLU(),
            nn.Linear(2, 1)
        )
    def forward(self, x):
        return self.net(x)

model = Model()
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 20000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(x_t)
    loss = criterion(outputs, y_t)
    loss.backward()
    optimizer.step()

    if epoch % 5000 == 0:
        print("Epoch", epoch, "Loss:", loss.item())
```

```python
# Predictions
pred_scaled = model(x_t).detach().numpy()
pred_real = scaler_y.inverse_transform(pred_scaled)

for i in range(10):
    print(f"Input: {x[i]} => Predicted: {pred_real[i][0]:.2f}, Target: {y[i][0]:.2f}")
```
